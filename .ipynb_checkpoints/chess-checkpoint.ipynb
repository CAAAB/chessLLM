```python
import chess
import chess.engine
import os
import sys # To check OS for clear screen command

# --- Configuration ---
# IMPORTANT: Replace this with the actual path to your Stockfish executable
# If you don't have/want Stockfish, analysis will be disabled.
STOCKFISH_PATH = "./stockfish-macos-m1-apple-silicon" # e.g., "C:/engines/stockfish_15_x64.exe" or "/usr/games/stockfish"

# Time limit for the engine's analysis (in seconds)
ANALYSIS_TIME_LIMIT = 2.0

# --- Helper Functions ---

def clear_screen():
    """Clears the terminal screen."""
    os.system('cls' if os.name == 'nt' else 'clear')

def display_board(board):
    """Prints the board and game status."""
    clear_screen()
    print("   Digital Twin Chess Board")
    print("-" * 28)
    # Simple text representation (could be enhanced)
    print(board)
    print("-" * 28)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.move_stack:
        print(f"Last move: {board.peek().uci()}") # Show last move in UCI
    print(f"Move count: {board.fullmove_number}")
    print("-" * 28)

def get_engine_analysis(board, engine, time_limit):
    """Gets the best move and evaluation from the engine."""
    if not engine:
        print("Analysis unavailable (Engine not configured or found).")
        return None, None

    try:
        print(f"\nAnalyzing (limit: {time_limit}s)...")
        # Analyze the current board position
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0] # Principal variation's first move
        score = info.get("score")

        if best_move and score is not None:
            # Convert score to a more understandable format
            pov_score = score.white() if board.turn == chess.WHITE else score.black()
            score_str = ""
            if pov_score.is_mate():
                score_str = f"Mate in {pov_score.mate()}"
            else:
                cp = pov_score.score(mate_score=10000) # Centipawns
                score_str = f"Score: {cp / 100.0:.2f}" # Convert to pawn units

            print(f"Engine Suggestion: {board.san(best_move)} ({score_str})")
            return board.san(best_move), score_str # Return SAN for readability
        else:
            print("Engine could not provide a suggestion.")
            return None, None

    except Exception as e:
        print(f"An error occurred during analysis: {e}")
        return None, None

def display_history(board):
    """Displays the move history in SAN."""
    print("\n--- Move History (SAN) ---")
    moves_san = []
    # Need to temporarily replay moves to get SAN
    temp_board = chess.Board()
    for i, move in enumerate(board.move_stack):
        move_san = temp_board.san(move)
        if temp_board.turn == chess.WHITE:
            moves_san.append(f"{temp_board.fullmove_number}. {move_san}")
        else:
            moves_san[-1] += f" {move_san}" # Append black's move to the last entry
        temp_board.push(move)

    if not moves_san:
        print("No moves made yet.")
    else:
        print("\n".join(moves_san))
    print("-" * 26)
    input("Press Enter to continue...")


# --- Main Game Logic ---

def main():
    board = chess.Board() # Initialize board to starting position
    engine = None

    # --- Initialize Engine ---
    print("Initializing Chess Twin...")
    try:
        if os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            print(f"Stockfish engine loaded from: {STOCKFISH_PATH}")
        else:
            print(f"Warning: Stockfish engine not found at '{STOCKFISH_PATH}'.")
            print("Download Stockfish: https://stockfishchess.org/download/")
            print("Analysis ('analyze' command) will be disabled.")
    except Exception as e:
        print(f"\nError initializing Stockfish engine: {e}")
        print("Analysis will be disabled.")
        engine = None # Ensure engine is None if setup failed

    input("Press Enter to start the game...")

    # --- Game Loop ---
    while not board.is_game_over(claim_draw=True): # Check for game end (checkmate, stalemate, draw rules)
        display_board(board)

        # Show potential checks
        if board.is_check():
            print("!! CHECK !!")

        # Prompt for move
        prompt = (f"Enter {'White' if board.turn == chess.WHITE else 'Black'}'s move "
                  f"(UCI format e.g., e2e4, g1f3, e7e8q for promotion),\n"
                  f"or type 'analyze', 'history', 'undo', 'result', 'quit': ")

        user_input = input(prompt).strip().lower()

        # --- Handle Commands ---
        if user_input == 'quit':
            print("Exiting game.")
            break
        elif user_input == 'analyze':
            get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
            input("Press Enter to continue...")
            continue # Ask for move again
        elif user_input == 'history':
            display_history(board)
            # display_board(board) # No need, display_history pauses
            continue # Ask for move again
        elif user_input == 'undo':
            if board.move_stack:
                board.pop() # Undo the last move
                print("Last move undone.")
                # If you want to undo a full turn (both white and black), pop again if possible
                # if board.move_stack:
                #    board.pop()
                #    print("Previous turn undone.")
            else:
                print("No moves to undo.")
            input("Press Enter to continue...")
            continue # Show updated board
        elif user_input == 'result':
            if board.is_game_over(claim_draw=True):
                 print(f"Current Result: {board.result(claim_draw=True)}")
                 print(f"Reason: {board.outcome(claim_draw=True)}")
            else:
                 print("Game is still in progress.")
            input("Press Enter to continue...")
            continue


        # --- Process Move Input ---
        try:
            # Try parsing as UCI (e.g., "e2e4", "e7e8q")
            move = board.parse_uci(user_input)
            if move in board.legal_moves:
                board.push(move) # Make the move on the digital board
            else:
                print(f"*** Illegal move: {user_input} ***")
                # Optionally list legal moves for debugging/help
                legal_moves_san = [board.san(m) for m in board.legal_moves]
                print(f"Legal moves are: {', '.join(legal_moves_san)}")
                input("Press Enter to try again...")

        except ValueError:
            # UCI parsing failed. Let's try common SAN (e.g., "Nf3", "e4", "O-O")
            # This is more complex due to ambiguity but user-friendlier
            try:
                move = board.parse_san(user_input)
                if move in board.legal_moves: # Double check legality (parse_san might return pseudo-legal)
                    board.push(move)
                else:
                     # This case should be rare if parse_san works but move is illegal
                     print(f"*** Illegal move (parsed as SAN): {user_input} ***")
                     input("Press Enter to try again...")
            except ValueError:
                 print(f"*** Invalid input: '{user_input}' ***")
                 print("Use UCI (e.g. e2e4, g1f3, a7a8q) or standard SAN (e.g. Nf3, e4, O-O).")
                 input("Press Enter to try again...")


    # --- Game Over ---
    print("\n" + "="*30)
    print("           GAME OVER")
    print("="*30)
    display_board(board) # Show final board state
    outcome = board.outcome(claim_draw=True)
    if outcome:
        print(f"Result: {outcome.result()}")
        print(f"Termination: {outcome.termination.name.replace('_', ' ').title()}")
    else:
        # Should not happen if loop condition is correct, but as fallback:
        print(f"Result: {board.result(claim_draw=True)}") # Get basic result string

    display_history(board) # Show full history at the end

    # --- Cleanup ---
    if engine:
        engine.quit()
        print("Stockfish engine closed.")

if __name__ == "__main__":
    main()
```

```python
# %%capture
# # Uncomment and run this cell if you need to install the library
# !pip install python-chess ipython # chess.svg is usually bundled

import chess
import chess.svg
import chess.engine
import random
import os
import sys
from IPython.display import display, clear_output, SVG, HTML

# --- Configuration ---
# IMPORTANT: Replace this with the actual path to your Stockfish executable
STOCKFISH_PATH = "./stockfish/stockfish-macos-m1-apple-silicon" # e.g., "C:/engines/stockfish_15_x64.exe" or "/usr/games/stockfish"

# Time limit for the engine's analysis (in seconds)
ANALYSIS_TIME_LIMIT = 2.0

# --- Helper Functions ---

def get_castling_rights_str(board):
    """Generates the standard castling rights string (KQkq or -)."""
    rights = []
    if board.has_kingside_castling_rights(chess.WHITE):
        rights.append("K")
    if board.has_queenside_castling_rights(chess.WHITE):
        rights.append("Q")
    if board.has_kingside_castling_rights(chess.BLACK):
        rights.append("k")
    if board.has_queenside_castling_rights(chess.BLACK):
        rights.append("q")

    return "".join(rights) if rights else "-"


def display_game(board, last_move=None, analysis_result=None, message=None):
    """Clears output and displays the SVG board, status, and messages."""
    clear_output(wait=True) # Prevents flickering

    # Generate SVG
    # Highlight the king if in check
    king_square = board.king(board.turn) if board.is_check() else None
    svg_board = chess.svg.board(
        board=board,
        lastmove=last_move,
        check=king_square, # Pass the square of the king in check
        coordinates=True, # Show board coordinates
        size=350 # Adjust size as needed
    )
    display(SVG(svg_board)) # Display SVG in Jupyter

    # Display Status
    print("-" * 30)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.is_check():
        print("!! CHECK !!")
    if last_move:
        try:
            # Use the board's internal move stack to get the SAN notation
            # This is more reliable than trying to re-parse the move object sometimes
            print(f"Last move: {board.san(last_move)}")
        except Exception as e:
            print(f"Last move: {last_move.uci()} (SAN error: {e})") # Fallback to UCI

    print(f"Move count: {board.fullmove_number}")
    castling_str = get_castling_rights_str(board)
    print(f"Castling Rights: {castling_str}") # Use the corrected helper function


    # Display any analysis results or messages
    if analysis_result:
        print("-" * 30)
        print(f"Engine Suggestion: {analysis_result[0]} ({analysis_result[1]})")
    if message:
        print("-" * 30)
        print(f"*** {message} ***")

    print("-" * 30)


def get_engine_analysis(board, engine, time_limit):
    """Gets the best move and evaluation from the engine."""
    if not engine:
        return None, "Analysis unavailable (Engine not configured or found)."

    try:
        print(f"Analyzing (limit: {time_limit}s)... Please wait.") # Give user feedback
        # Analyze the current board position
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0] # Principal variation's first move
        score = info.get("score")

        if best_move and score is not None:
            pov_score = score.white() if board.turn == chess.WHITE else score.black()
            score_str = ""
            if pov_score.is_mate():
                mate_in = pov_score.mate()
                 # Ensure mate_in is not None and handle negative values for black's perspective if needed
                score_str = f"Mate in {abs(mate_in)}" if mate_in is not None else "Mate (score unavailable)"
            else:
                cp = pov_score.score(mate_score=10000) # Centipawns
                score_str = f"Score: {cp / 100.0:.2f}" if cp is not None else "Score: N/A"

            # Ensure we can get SAN for the suggested move
            try:
                best_move_san = board.san(best_move)
            except Exception:
                best_move_san = best_move.uci() # Fallback to UCI if SAN fails

            return best_move_san, score_str # Return SAN for readability
        else:
            return None, "Engine could not provide a suggestion."

    except Exception as e:
        print(f"An error occurred during analysis: {e}") # Show error in console
        return None, f"Analysis error: {e}"


def display_history(board):
    """Displays the move history in SAN."""
    history_html = "<h3>Move History (SAN)</h3><ol>"
    # Need to temporarily replay moves to get SAN
    temp_board = chess.Board(chess960=board.chess960)
    try:
        # Use the initial FEN from the root node of the game tree
        temp_board.set_fen(board.root().fen())
    except Exception as e:
        # Fallback if root node access fails (shouldn't usually happen)
        print(f"Error getting start FEN for history: {e}")
        history_html += "<li>Error generating history</li></ol>"
        display(HTML(history_html))
        return


    full_move_parts = [] # Store parts like "1. e4", "e5"
    for i, move in enumerate(board.move_stack):
        try:
            move_san = temp_board.san(move)
        except Exception:
            move_san = move.uci() # Fallback if SAN fails

        if temp_board.turn == chess.WHITE:
             # Start a new list item for the full move number
             if full_move_parts: # Add the completed previous move pair
                 history_html += f"<li>{' '.join(full_move_parts)}</li>"
             full_move_parts = [f"{temp_board.fullmove_number}. {move_san}"]
        else:
            # Append black's move
            full_move_parts.append(move_san)


        try:
            temp_board.push(move) # Make move on temp board
        except Exception as e:
            history_html += f"<li>Error processing move {move.uci()}: {e}</li>"
            break # Stop processing if a move fails

    if full_move_parts: # Add the last move pair/single move
        history_html += f"<li>{' '.join(full_move_parts)}</li>"


    if not board.move_stack:
        history_html += "<li>No moves made yet.</li>"

    history_html += "</ol>"
    display(HTML(history_html))


# --- Main Game Logic ---

def play_chess960():
    board = chess.Board(chess960=True) # Enable Chess960 mode
    engine = None
    last_move_made = None
    analysis_result = None # Store last analysis to redisplay
    message = None # Store messages (like errors) to redisplay
    start_pos_num = -1 # Initialize

    # --- Setup Chess960 Position ---
    while True:
        clear_output()
        pos_choice = input("Enter Chess960 position number (0-959), 'random', or leave blank for random: ").strip().lower()
        if not pos_choice or pos_choice == 'random':
            start_pos_num = random.randint(0, 959) # Includes standard
            break
        else:
            try:
                start_pos_num = int(pos_choice)
                if 0 <= start_pos_num <= 959:
                    break
                else:
                    print("Position number must be between 0 and 959.")
                    input("Press Enter to try again...") # Pause for user to see msg
            except ValueError:
                print("Invalid input. Please enter a number or 'random'.")
                input("Press Enter to try again...") # Pause for user to see msg

    board.set_chess960_pos(start_pos_num)
    start_fen = board.fen() # Store the starting FEN
    print(f"Starting Chess960 Position {start_pos_num}")
    print(f"Starting FEN: {start_fen}")
    input("Press Enter to continue...")


    # --- Initialize Engine ---
    try:
        if STOCKFISH_PATH and os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            # Configure engine for Chess960 (most modern engines detect via FEN)
            # Try setting the option if available, ignore error if not supported
            try:
                 engine.configure({"UCI_Chess960": True})
                 print("Set UCI_Chess960 option for engine.")
            except chess.engine.EngineError:
                 print("Engine does not support UCI_Chess960 option (usually fine, detects via FEN).")
            print(f"Stockfish engine loaded from: {STOCKFISH_PATH}")
        else:
            print(f"Warning: Stockfish engine not found or path not set '{STOCKFISH_PATH}'.")
            print("Analysis ('analyze' command) will be disabled.")
    except Exception as e:
        print(f"\nError initializing Stockfish engine: {e}")
        print("Analysis will be disabled.")
        engine = None

    input("Press Enter to start the game...")

    # --- Game Loop ---
    while True: # Loop indefinitely until break
        display_game(board, last_move=last_move_made, analysis_result=analysis_result, message=message)
        message = None # Clear message after displaying
        analysis_result = None # Clear analysis after displaying

        # Check for game over *before* asking for input
        if board.is_game_over(claim_draw=True):
            message = "Game Over!"
            display_game(board, last_move=last_move_made, analysis_result=analysis_result, message=message) # Show final state immediately
            break # Exit the game loop

        # Prompt for move
        prompt = (f"Enter move (UCI: e.g., e2e4 | SAN: e.g., Nf3, O-O)\n"
                  f"Cmds: analyze, history, undo, fen, pos, result, quit: ")

        user_input = input(prompt).strip()

        # --- Handle Commands ---
        cmd = user_input.lower()
        if cmd == 'quit':
            print("Exiting game.")
            break
        elif cmd == 'analyze':
            suggestion, score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
            if suggestion:
                analysis_result = (suggestion, score)
            else:
                message = score # Display error message if analysis failed
            last_move_made = None # Clear last move highlight after command
            continue # Redisplay board with analysis/message
        elif cmd == 'history':
            # Keep current display, add history below, then prompt
            display_history(board)
            input("Press Enter to return to the game...")
            last_move_made = None
            continue # Need to redisplay the board cleanly
        elif cmd == 'undo':
            if board.move_stack:
                try:
                    board.pop()
                    message = "Last move undone."
                    # Update last_move_made to highlight the *new* last move
                    last_move_made = board.peek() if board.move_stack else None
                except IndexError:
                     message = "Cannot undo." # Should not happen if move_stack is checked
                     last_move_made = None
            else:
                message = "No moves to undo."
                last_move_made = None
            continue # Show updated board
        elif cmd == 'fen':
             message = f"Current FEN: {board.fen()}"
             last_move_made = None
             continue
        elif cmd == 'pos':
             message = f"Starting Position: {start_pos_num} | FEN: {start_fen}"
             last_move_made = None
             continue
        elif cmd == 'result':
             # This command is less useful now game over is checked above, but keep it
             if board.is_game_over(claim_draw=True):
                 res_info = board.outcome(claim_draw=True)
                 term = res_info.termination.name.replace('_', ' ').title()
                 message = f"Result: {res_info.result()} ({term})"
             else:
                 message = "Game is still in progress."
             last_move_made = None
             continue

        # --- Process Move Input ---
        move = None
        try:
            # Prioritize UCI since it's less ambiguous
            move = board.parse_uci(user_input)
        except ValueError:
            # If UCI fails, try SAN (more user-friendly but can be ambiguous)
            try:
                # For Chess960 castling use O-O or O-O-O
                # The library maps these SAN moves to the correct squares
                move = board.parse_san(user_input)
            except ValueError as e_san:
                 # Check for specific ambiguous/illegal move errors from parse_san
                 err_str = str(e_san).lower()
                 if "ambiguous" in err_str:
                     message = f"Ambiguous move: '{user_input}'. Specify piece/file/rank (e.g. Nbe2, R1a3)."
                 elif "illegal" in err_str:
                      message = f"Illegal move: '{user_input}'."
                 else:
                      message = f"Invalid input: '{user_input}'. Use UCI (e.g. e2e4) or SAN (e.g. Nf3)."
                 last_move_made = None
                 continue # Ask for input again

        # Now, double-check if the successfully parsed move is actually legal
        if move in board.legal_moves:
            board.push(move)
            last_move_made = move # Store the move for highlighting
            analysis_result = None # Clear old analysis after a valid move
            message = None # Clear any previous error messages
        else:
            # Move was parsed but is not legal (e.g., leaves king in check)
            parsed_as = user_input # Default to user input
            try:
                # Try to show the SAN of the illegal move attempt for clarity
                parsed_as = board.san(move)
            except ValueError:
                 try:
                     # Fallback to UCI if SAN fails
                     parsed_as = move.uci()
                 except AttributeError:
                      pass # Keep original input if move object is bad


            reason = "puts king in check" if board.is_pseudo_legal(move) else "is not legal"
            message = f"Illegal move: {user_input} (parsed {parsed_as}) - {reason}."

            # Optionally list legal moves (can be very long!)
            # legal_moves_san = [board.san(m) for m in board.legal_moves]
            # print(f"\nLegal moves: {', '.join(legal_moves_san)}") # Display below prompt maybe

            last_move_made = None # Clear highlight on error


    # --- Game Over Sequence (already broken out of loop) ---
    print("\n" + "="*30)
    print("           GAME OVER")
    print("="*30)
    # Final display already happened before break, but we can add result text
    outcome = board.outcome(claim_draw=True)
    if outcome:
        termination_reason = outcome.termination.name.replace('_', ' ').title()
        print(f"\nResult: {outcome.result()} ({termination_reason})")
    else:
        print(f"\nResult: {board.result(claim_draw=True)}") # Fallback

    display_history(board) # Show full history at the end

    # --- Cleanup ---
    if engine:
        try:
            engine.quit()
            print("\nStockfish engine closed.")
        except chess.engine.EngineTerminatedError:
            print("\nStockfish engine already terminated.")
        except Exception as e:
             print(f"\nError closing Stockfish engine: {e}")


# --- Start the Game ---
play_chess960()
```

```python
import chess
import chess.svg
import chess.engine
import random
import os
import sys
import math # For clamping mate scores potentially
from IPython.display import display, clear_output, SVG, HTML

# --- Configuration ---
# IMPORTANT: Replace this with the actual path to your Stockfish executable
STOCKFISH_PATH = "./stockfish/stockfish-macos-m1-apple-silicon" # e.g., "/usr/games/stockfish" or "C:/engines/stockfish.exe"

# Time limit for the engine's analysis (in seconds)
ANALYSIS_TIME_LIMIT = 2.0

# Evaluation bar parameters
EVAL_BAR_HEIGHT = 350 # px, match board size for alignment
EVAL_BAR_WIDTH = 20   # px
# Score range to map to the bar (+/- this value in centipawns)
# Scores outside this range will be clamped to the ends of the bar.
EVAL_CLAMP_LIMIT = 800 # +/- 8 pawns default

# --- Helper Functions ---

def get_castling_rights_str(board):
    """Generates the standard castling rights string (KQkq or -)."""
    rights = []
    if board.has_kingside_castling_rights(chess.WHITE):
        rights.append("K")
    if board.has_queenside_castling_rights(chess.WHITE):
        rights.append("Q")
    if board.has_kingside_castling_rights(chess.BLACK):
        rights.append("k")
    if board.has_queenside_castling_rights(chess.BLACK):
        rights.append("q")
    return "".join(rights) if rights else "-"

def score_to_white_percentage(score, clamp_limit=EVAL_CLAMP_LIMIT):
    """
    Converts a chess.engine.Score object to a percentage (0-100) for the white part of the eval bar.
    Handles centipawns and mate scores. Clamps centipawn scores.
    Returns None if score is invalid.
    """
    if score is None:
        return None

    pov_score = score.white() # Get score from White's perspective

    if pov_score.is_mate():
        mate_val = pov_score.mate()
        if mate_val is not None:
            return 100 if mate_val > 0 else 0 # 100% for white mate, 0% for black mate
        else:
             return None # Mate score but value is None? Shouldn't happen often

    # Handle centipawn scores
    cp = pov_score.score(mate_score=clamp_limit * 2) # Provide a large mate_score to prefer cp
    if cp is None:
        return None # No valid centipawn score

    # Clamp the score to the defined limits
    clamped_cp = max(-clamp_limit, min(clamp_limit, cp))

    # Normalize the clamped score to a 0-1 range
    # -clamp_limit maps to 0, +clamp_limit maps to 1
    normalized = (clamped_cp + clamp_limit) / (2 * clamp_limit)

    # Convert to percentage
    return normalized * 100


def create_eval_bar_html(white_percentage, height=EVAL_BAR_HEIGHT, width=EVAL_BAR_WIDTH):
    """Generates HTML/CSS for the evaluation bar."""
    if white_percentage is None:
        # Maybe return a neutral gray bar or nothing? Let's return empty for now.
        # Could also return a 50/50 bar: return create_eval_bar_html(50, height, width)
         return ""

    # Ensure percentage is within 0-100
    white_percentage = max(0, min(100, white_percentage))
    black_percentage = 100 - white_percentage

    # Style for the container (defines size, border, and acts as the black part)
    container_style = (
        f"width: {width}px; "
        f"height: {height}px; "
        f"background-color: #333; " # Black/dark gray background
        f"border: 1px solid #666; " # Optional border
        f"position: relative; "      # Needed for absolute positioning of the white part
        f"overflow: hidden;"         # Hide overflow
        f"box-sizing: border-box;"   # Include border in width/height
    )

    # Style for the inner white div
    white_style = (
        f"width: 100%; "
        f"height: {white_percentage}%; "
        f"background-color: white; "
        f"position: absolute; "
        f"bottom: 0; "
        f"left: 0;"
    )

    # Combine into HTML
    html = (
        f'<div style="{container_style}">'
        f'  <div style="{white_style}"></div>'
        f'</div>'
    )
    return html

def display_game(board, last_move=None, analysis_info=None, message=None):
    """Clears output and displays the SVG board, status, optional eval bar, and messages."""
    clear_output(wait=True)

    # --- Prepare Board SVG ---
    king_square = board.king(board.turn) if board.is_check() else None
    svg_board_str = chess.svg.board(
        board=board,
        lastmove=last_move,
        check=king_square,
        coordinates=True,
        size=EVAL_BAR_HEIGHT # Match board height to eval bar height
    )

    # --- Prepare Eval Bar HTML (if analysis info available) ---
    eval_bar_html = ""
    raw_score = None
    analysis_text = "" # Text suggestion for display below

    if analysis_info:
        best_move_san, score_str, raw_score = analysis_info # Unpack the tuple
        analysis_text = f"Engine Suggestion: {best_move_san} ({score_str})"
        white_percentage = score_to_white_percentage(raw_score, EVAL_CLAMP_LIMIT)
        if white_percentage is not None: # Only create bar if score is valid
            eval_bar_html = create_eval_bar_html(white_percentage, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)
        else:
             # Optional: show a placeholder if score was expected but invalid
             # eval_bar_html = create_eval_bar_html(50, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH) # 50/50 bar
             pass # Or just show nothing


    # --- Combine Board and Eval Bar using HTML Table ---
    # Use valign="top" to align items to the top of their cells
    combined_html = f"""
    <table style="border-collapse: collapse; border: none;">
        <tr>
            <td style="padding: 0 10px 0 0; vertical-align: top;">{eval_bar_html}</td>
            <td style="padding: 0; vertical-align: top;">{svg_board_str}</td>
        </tr>
    </table>
    """
    display(HTML(combined_html))

    # --- Display Status Text Below ---
    print("-" * 30)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.is_check():
        print("!! CHECK !!")
    if last_move:
        try:
            print(f"Last move: {board.san(last_move)}")
        except Exception as e:
            print(f"Last move: {last_move.uci()} (SAN error: {e})")

    print(f"Move count: {board.fullmove_number}")
    castling_str = get_castling_rights_str(board)
    print(f"Castling Rights: {castling_str}")

    if analysis_text:
        print("-" * 30)
        print(analysis_text) # Display the text part of the analysis result

    if message:
        print("-" * 30)
        print(f"*** {message} ***")

    print("-" * 30)


def get_engine_analysis(board, engine, time_limit):
    """
    Gets the best move (SAN), score (string), and raw score object from the engine.
    Returns (None, "Error Message", None) on failure.
    """
    if not engine:
        return None, "Analysis unavailable (Engine not configured or found).", None

    try:
        print(f"Analyzing (limit: {time_limit}s)... Please wait.")
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0]
        score = info.get("score") # This is the raw score object

        if best_move and score is not None:
            pov_score = score.white() if board.turn == chess.WHITE else score.black()
            score_str = ""
            if pov_score.is_mate():
                mate_in = pov_score.mate()
                score_str = f"Mate in {abs(mate_in)}" if mate_in is not None else "Mate (N/A)"
            else:
                cp = pov_score.score(mate_score=EVAL_CLAMP_LIMIT * 2)
                score_str = f"Score: {cp / 100.0:.2f}" if cp is not None else "Score: N/A"

            try:
                best_move_san = board.san(best_move)
            except Exception:
                best_move_san = best_move.uci() # Fallback

            return best_move_san, score_str, score # <<< Return the raw score object too

        else:
            return None, "Engine could not provide a suggestion.", None

    except Exception as e:
        print(f"An error occurred during analysis: {e}")
        return None, f"Analysis error: {e}", None


def display_history(board):
    """Displays the move history in SAN. (No changes needed here)"""
    history_html = "<h3>Move History (SAN)</h3><ol>"
    temp_board = chess.Board(chess960=board.chess960)
    try:
        temp_board.set_fen(board.root().fen())
    except Exception as e:
        print(f"Error getting start FEN for history: {e}")
        history_html += "<li>Error generating history</li></ol>"
        display(HTML(history_html))
        return

    full_move_parts = []
    for i, move in enumerate(board.move_stack):
        try:
            move_san = temp_board.san(move)
        except Exception:
            move_san = move.uci() # Fallback if SAN fails

        if temp_board.turn == chess.WHITE:
             if full_move_parts:
                 history_html += f"<li>{' '.join(full_move_parts)}</li>"
             full_move_parts = [f"{temp_board.fullmove_number}. {move_san}"]
        else:
            full_move_parts.append(move_san)

        try:
            temp_board.push(move)
        except Exception as e:
            history_html += f"<li>Error processing move {move.uci()}: {e}</li>"
            break

    if full_move_parts:
        history_html += f"<li>{' '.join(full_move_parts)}</li>"

    if not board.move_stack:
        history_html += "<li>No moves made yet.</li>"

    history_html += "</ol>"
    display(HTML(history_html))


# --- Main Game Logic ---

def play_chess960():
    board = chess.Board(chess960=True)
    engine = None
    last_move_made = None
    analysis_info = None # Store tuple: (best_move_san, score_str, raw_score)
    message = None
    start_pos_num = -1

    # --- Setup Chess960 Position ---
    while True:
        clear_output()
        pos_choice = input("Enter Chess960 position number (0-959), 'random', or leave blank for random: ").strip().lower()
        if not pos_choice or pos_choice == 'random':
            start_pos_num = random.randint(0, 959)
            break
        else:
            try:
                start_pos_num = int(pos_choice)
                if 0 <= start_pos_num <= 959:
                    break
                else:
                    print("Position number must be between 0 and 959.")
                    input("Press Enter to try again...")
            except ValueError:
                print("Invalid input. Please enter a number or 'random'.")
                input("Press Enter to try again...")

    board.set_chess960_pos(start_pos_num)
    start_fen = board.fen()
    print(f"Starting Chess960 Position {start_pos_num}")
    print(f"Starting FEN: {start_fen}")
    input("Press Enter to continue...")

    # --- Initialize Engine ---
    try:
        if STOCKFISH_PATH and os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            try:
                 engine.configure({"UCI_Chess960": True})
                 print("Set UCI_Chess960 option for engine.")
            except chess.engine.EngineError:
                 print("Engine does not support UCI_Chess960 option (usually fine, detects via FEN).")
            print(f"Stockfish engine loaded from: {STOCKFISH_PATH}")
        else:
            print(f"Warning: Stockfish engine not found or path not set '{STOCKFISH_PATH}'.")
            print("Analysis ('analyze' command) will be disabled.")
    except Exception as e:
        print(f"\nError initializing Stockfish engine: {e}")
        print("Analysis will be disabled.")
        engine = None

    input("Press Enter to start the game...")

    # --- Game Loop ---
    while True:
        # Display using the combined function (handles eval bar if analysis_info exists)
        display_game(board, last_move=last_move_made, analysis_info=analysis_info, message=message)
        message = None        # Clear message after displaying
        analysis_info = None  # <<<< IMPORTANT: Clear analysis info after display
                               # So eval bar only shows right after 'analyze' command

        # Check for game over *before* asking for input
        if board.is_game_over(claim_draw=True):
            message = "Game Over!"
            # Show final state WITH the message, but clear any lingering analysis bar
            display_game(board, last_move=last_move_made, analysis_info=None, message=message)
            break # Exit the game loop

        # Prompt for move
        prompt = (f"Enter move (UCI: e.g., e2e4 | SAN: e.g., Nf3, O-O)\n"
                  f"Cmds: analyze, history, undo, fen, pos, result, quit: ")
        user_input = input(prompt).strip()

        # --- Handle Commands ---
        cmd = user_input.lower()
        if cmd == 'quit':
            print("Exiting game.")
            break
        elif cmd == 'analyze':
            # Store the complete analysis result (including raw score)
            best_move_san, score_str, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
            if best_move_san: # Check if analysis was successful
                analysis_info = (best_move_san, score_str, raw_score)
            else:
                message = score_str # Display error message if analysis failed
            last_move_made = None # Clear last move highlight after command
            continue # Redisplay board (will now include eval bar if analysis succeeded)
        elif cmd == 'history':
            display_history(board)
            input("Press Enter to return to the game...")
            last_move_made = None
            continue
        elif cmd == 'undo':
            if board.move_stack:
                try:
                    board.pop()
                    message = "Last move undone."
                    last_move_made = board.peek() if board.move_stack else None
                except IndexError:
                     message = "Cannot undo."
                     last_move_made = None
            else:
                message = "No moves to undo."
                last_move_made = None
            # Clear any lingering analysis after undo
            analysis_info = None
            continue # Show updated board
        elif cmd == 'fen':
             message = f"Current FEN: {board.fen()}"
             last_move_made = None
             continue
        elif cmd == 'pos':
             message = f"Starting Position: {start_pos_num} | FEN: {start_fen}"
             last_move_made = None
             continue
        elif cmd == 'result':
             if board.is_game_over(claim_draw=True):
                 res_info = board.outcome(claim_draw=True)
                 term = res_info.termination.name.replace('_', ' ').title()
                 message = f"Result: {res_info.result()} ({term})"
             else:
                 message = "Game is still in progress."
             last_move_made = None
             continue

        # --- Process Move Input ---
        move = None
        try:
            move = board.parse_uci(user_input)
        except ValueError:
            try:
                move = board.parse_san(user_input)
            except ValueError as e_san:
                 err_str = str(e_san).lower()
                 if "ambiguous" in err_str:
                     message = f"Ambiguous move: '{user_input}'. Specify piece/file/rank (e.g. Nbe2, R1a3)."
                 elif "illegal" in err_str:
                      message = f"Illegal move: '{user_input}'."
                 else:
                      message = f"Invalid input: '{user_input}'. Use UCI (e.g. e2e4) or SAN (e.g. Nf3)."
                 last_move_made = None
                 continue

        if move in board.legal_moves:
            board.push(move)
            last_move_made = move
            analysis_info = None # Clear analysis after a valid move is made
            message = None
        else:
            parsed_as = user_input
            try: parsed_as = board.san(move)
            except ValueError:
                 try: parsed_as = move.uci()
                 except AttributeError: pass

            reason = "puts king in check" if board.is_pseudo_legal(move) else "is not legal"
            message = f"Illegal move: {user_input} (parsed {parsed_as}) - {reason}."
            last_move_made = None

    # --- Game Over Sequence ---
    print("\n" + "="*30)
    print("           GAME OVER")
    print("="*30)
    # Final display happened before break, add result text below it
    outcome = board.outcome(claim_draw=True)
    if outcome:
        termination_reason = outcome.termination.name.replace('_', ' ').title()
        print(f"\nResult: {outcome.result()} ({termination_reason})")
    else:
        print(f"\nResult: {board.result(claim_draw=True)}") # Fallback

    display_history(board)

    # --- Cleanup ---
    if engine:
        try:
            engine.quit()
            print("\nStockfish engine closed.")
        except chess.engine.EngineTerminatedError:
            print("\nStockfish engine already terminated.")
        except Exception as e:
             print(f"\nError closing Stockfish engine: {e}")


# --- Start the Game ---
play_chess960()
```

```python
# %%capture
# # Uncomment and run this cell if you need to install the libraries
# !pip install python-chess ipython matplotlib

import chess
import chess.svg
import chess.engine
import random
import os
import sys
import math
import io                     # For saving plot to memory
import base64               # For embedding plot image
import matplotlib.pyplot as plt
from IPython.display import display, clear_output, SVG, HTML

# --- Configuration ---
STOCKFISH_PATH = "./stockfish/stockfish-macos-m1-apple-silicon" # ADJUST THIS PATH
ANALYSIS_TIME_LIMIT = 1.0 # Reduced for faster auto-analysis
EVAL_BAR_HEIGHT = 350
EVAL_BAR_WIDTH = 20
EVAL_CLAMP_LIMIT = 800 # +/- 8 pawns for bar range
PLOT_HEIGHT = 150      # Height of the plot image in pixels
PLOT_WIDTH = EVAL_BAR_HEIGHT + EVAL_BAR_WIDTH + 10 # Make plot width match board+bar approx
# Value to represent mate scores on the plot (+/-)
# Needs to be larger than EVAL_CLAMP_LIMIT * 100 (in centipawns)
MATE_SCORE_PLOT_VALUE = 15000

# --- Helper Functions ---

def get_castling_rights_str(board):
    rights = []
    if board.has_kingside_castling_rights(chess.WHITE): rights.append("K")
    if board.has_queenside_castling_rights(chess.WHITE): rights.append("Q")
    if board.has_kingside_castling_rights(chess.BLACK): rights.append("k")
    if board.has_queenside_castling_rights(chess.BLACK): rights.append("q")
    return "".join(rights) if rights else "-"

def score_to_white_percentage(score, clamp_limit=EVAL_CLAMP_LIMIT):
    if score is None: return None
    pov_score = score.white()
    if pov_score.is_mate():
        mate_val = pov_score.mate()
        return 100 if mate_val is not None and mate_val > 0 else 0
    cp = pov_score.score(mate_score=clamp_limit * 2)
    if cp is None: return None
    clamped_cp = max(-clamp_limit, min(clamp_limit, cp))
    normalized = (clamped_cp + clamp_limit) / (2 * clamp_limit)
    return normalized * 100

def score_to_numerical(score):
    """Converts score to centipawns (White POV), using large values for mate."""
    if score is None: return 0 # Treat unknown score as neutral for plot history
    pov_score = score.white()
    if pov_score.is_mate():
        mate_val = pov_score.mate()
        if mate_val is not None:
            return MATE_SCORE_PLOT_VALUE if mate_val > 0 else -MATE_SCORE_PLOT_VALUE
        else:
            return 0 # Undefined mate
    # Use a large mate_score to prefer cp, default to 0 if cp is None
    cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE * 2)
    return cp if cp is not None else 0

def create_eval_bar_html(white_percentage, height=EVAL_BAR_HEIGHT, width=EVAL_BAR_WIDTH):
    if white_percentage is None: return ""
    white_percentage = max(0, min(100, white_percentage))
    container_style = f"width: {width}px; height: {height}px; background-color: #333; border: 1px solid #666; position: relative; overflow: hidden; box-sizing: border-box;"
    white_style = f"width: 100%; height: {white_percentage}%; background-color: white; position: absolute; bottom: 0; left: 0;"
    return f'<div style="{container_style}"><div style="{white_style}"></div></div>'

def create_eval_plot_html(eval_history, plot_width_px=PLOT_WIDTH, plot_height_px=PLOT_HEIGHT):
    """Generates HTML containing a base64 encoded plot of the evaluation history."""
    if not eval_history:
        # Return an empty placeholder matching the plot size? Or just empty string.
        # return f'<div style="width:{plot_width_px}px; height:{plot_height_px}px; border: 1px dashed lightgrey; display: flex; align-items: center; justify-content: center; color: grey;">Evaluation Plot Area</div>'
        return "" # Return empty if no history


    plies = [item[0] for item in eval_history]
    scores = [item[1] / 100.0 for item in eval_history] # Convert cp to pawns for display

    # Determine plot y-axis limits dynamically but capped
    max_abs_score = 0
    if scores:
        max_abs_score = max(abs(s) for s in scores if abs(s) < MATE_SCORE_PLOT_VALUE / 100.0) # Ignore obvious mates for limits
        max_abs_score = max(max_abs_score, 2) # Ensure at least +/- 2 range

    # Cap the y-limit, slightly above the clamp limit for the bar
    plot_y_limit = min(max(max_abs_score * 1.2, 2), EVAL_CLAMP_LIMIT / 100.0 * 1.5) # Up to 1.5x bar limit

    # Handle Mate scores for display clipping
    plot_y_limit_final = MATE_SCORE_PLOT_VALUE / 100.0 # Use the large value for actual clipping if mate occurs

    fig, ax = plt.subplots(figsize=(plot_width_px / 80, plot_height_px / 80), dpi=80) # Adjust figsize based on desired pixel dimensions
    ax.plot(plies, scores, marker='.', linestyle='-', color='blue')
    ax.axhline(0, color='black', linewidth=0.5, linestyle='--') # Zero line

    ax.set_xlabel("Ply (Half-moves)")
    ax.set_ylabel("Eval (Pawns)")
    ax.set_title("Evaluation History (White's POV)")
    ax.grid(True, linestyle=':', linewidth=0.5)

    # Set y-limits, clipping extreme mate scores visually
    ax.set_ylim(-plot_y_limit_final, plot_y_limit_final) # Clip extreme values

    # Add text annotations near y=0 for clarity if space permits
    # Add plot limits text
    ax.text(plies[-1] if plies else 0, plot_y_limit * 0.9, f'+{plot_y_limit:.1f}', verticalalignment='top', horizontalalignment='right', fontsize=8, color='grey')
    ax.text(plies[-1] if plies else 0, -plot_y_limit * 0.9, f'-{plot_y_limit:.1f}', verticalalignment='bottom', horizontalalignment='right', fontsize=8, color='grey')


    # Save plot to a BytesIO object
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight') # Use bbox_inches='tight'
    plt.close(fig) # Close the plot to free memory
    buf.seek(0)

    # Encode image to base64
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    buf.close()

    # Create HTML img tag
    return f'<img src="data:image/png;base64,{img_base64}" alt="Evaluation Plot"/>'


def display_game(board, meter_visible, eval_history, last_raw_score, last_move=None, analysis_text=None, message=None):
    """Displays board, optional meter/plot, status, and messages."""
    clear_output(wait=True)

    # --- Prepare Board SVG ---
    king_square = board.king(board.turn) if board.is_check() else None
    svg_board_str = chess.svg.board(
        board=board, lastmove=last_move, check=king_square,
        coordinates=True, size=EVAL_BAR_HEIGHT
    )

    # --- Prepare Eval Bar and Plot (if meter visible) ---
    eval_bar_html = ""
    eval_plot_html = ""
    if meter_visible:
        # Use the last known score for the bar when meter is on
        if last_raw_score:
            white_percentage = score_to_white_percentage(last_raw_score, EVAL_CLAMP_LIMIT)
            eval_bar_html = create_eval_bar_html(white_percentage, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)
        else: # Show neutral bar if no score yet
            eval_bar_html = create_eval_bar_html(50, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)

        # Create plot from history
        eval_plot_html = create_eval_plot_html(eval_history, PLOT_WIDTH, PLOT_HEIGHT)

    # --- Combine Board, Bar, and Plot using HTML ---
    # Eval bar left, board right, plot below spanning both (or just below board if no bar)
    combined_html = f"""
    <table style="border-collapse: collapse; border: none;">
        <tr>
            <td style="padding: 0 10px 0 0; vertical-align: top;">{eval_bar_html}</td>
            <td style="padding: 0; vertical-align: top;">{svg_board_str}</td>
        </tr>
        """
    # Add plot row only if it was generated
    if eval_plot_html:
        combined_html += f"""
        <tr>
            <td colspan="2" style="padding-top: 10px; text-align: left;">{eval_plot_html}</td>
        </tr>
        """
    combined_html += "</table>"
    display(HTML(combined_html))


    # --- Display Status Text Below ---
    print("-" * 30)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.is_check(): print("!! CHECK !!")
    if last_move:
        try: print(f"Last move: {board.san(last_move)}")
        except Exception as e: print(f"Last move: {last_move.uci()} (SAN error: {e})")
    print(f"Move count: {board.fullmove_number}")
    print(f"Castling Rights: {get_castling_rights_str(board)}")
    print(f"Ply: {len(board.move_stack)}") # Show ply count

    if analysis_text: # Display one-off analysis text if present
        print("-" * 30)
        print(analysis_text)

    if message:
        print("-" * 30)
        print(f"*** {message} ***")
    print("-" * 30)


def get_engine_analysis(board, engine, time_limit):
    """Returns (best_move_san, score_str, raw_score) or (None, "Error", None)."""
    if not engine:
        return None, "Analysis unavailable (Engine not configured).", None
    try:
        # print(f"Analyzing (limit: {time_limit}s)...") # Can be noisy for auto-analysis
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0]
        score = info.get("score")
        if best_move and score is not None:
            pov_score = score.white()
            score_str = ""
            if pov_score.is_mate():
                mate_in = pov_score.mate()
                score_str = f"Mate in {abs(mate_in)}" if mate_in is not None else "Mate (N/A)"
            else:
                cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE * 2)
                score_str = f"Score: {cp / 100.0:.2f}" if cp is not None else "Score: N/A"
            try: best_move_san = board.san(best_move)
            except Exception: best_move_san = best_move.uci()
            return best_move_san, score_str, score
        else:
            return None, "Engine could not provide suggestion.", None
    except Exception as e:
        print(f"Analysis error: {e}") # Show error in console
        return None, f"Analysis error: {e}", None

def display_history(board):
    """Displays the move history in SAN."""
    history_html = "<h3>Move History (SAN)</h3><ol>"
    temp_board = chess.Board(chess960=board.chess960)
    try: temp_board.set_fen(board.root().fen())
    except Exception as e:
        print(f"Error getting start FEN for history: {e}")
        display(HTML("<li>Error generating history</li></ol>"))
        return

    full_move_parts = []
    for i, move in enumerate(board.move_stack):
        try: move_san = temp_board.san(move)
        except Exception: move_san = move.uci()
        if temp_board.turn == chess.WHITE:
             if full_move_parts: history_html += f"<li>{' '.join(full_move_parts)}</li>"
             full_move_parts = [f"{temp_board.fullmove_number}. {move_san}"]
        else:
            full_move_parts.append(move_san)
        try: temp_board.push(move)
        except Exception as e:
            history_html += f"<li>Error processing move {move.uci()}: {e}</li>"
            break
    if full_move_parts: history_html += f"<li>{' '.join(full_move_parts)}</li>"
    if not board.move_stack: history_html += "<li>No moves made yet.</li>"
    history_html += "</ol>"
    display(HTML(history_html))

# --- Main Game Logic ---
def play_chess960():
    board = chess.Board(chess960=True)
    engine = None
    last_move_made = None
    message = None
    start_pos_num = -1

    # --- State for Meter and Plot ---
    meter_visible = False
    eval_history = [] # List of tuples: (ply, score_numerical_cp)
    last_raw_score = None # Store the last raw score object for the bar
    one_off_analysis_text = None # For displaying result of 'analyze' cmd

    # --- Setup Chess960 Position ---
    # ... (identical setup code as before) ...
    while True:
        clear_output()
        pos_choice = input("Enter Chess960 position number (0-959), 'random', or leave blank for random: ").strip().lower()
        if not pos_choice or pos_choice == 'random':
            start_pos_num = random.randint(0, 959)
            break
        else:
            try:
                start_pos_num = int(pos_choice)
                if 0 <= start_pos_num <= 959: break
                else: print("Position number must be between 0 and 959."); input("Press Enter...")
            except ValueError: print("Invalid input."); input("Press Enter...")
    board.set_chess960_pos(start_pos_num)
    start_fen = board.fen()
    # Add initial position score (0) if meter starts enabled (optional)
    # if meter_visible: eval_history.append((0, 0)); last_raw_score = chess.engine.Score(0)
    print(f"Starting Chess960 Position {start_pos_num}")
    print(f"Starting FEN: {start_fen}")
    input("Press Enter to continue...")

    # --- Initialize Engine ---
    # ... (identical engine init code as before) ...
    try:
        if STOCKFISH_PATH and os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            try: engine.configure({"UCI_Chess960": True}); print("Set UCI_Chess960 option.")
            except chess.engine.EngineError: print("UCI_Chess960 option not supported (usually ok).")
            print(f"Stockfish engine loaded: {STOCKFISH_PATH}")
        else: print(f"Warning: Stockfish not found ('{STOCKFISH_PATH}'). Analysis disabled.")
    except Exception as e: print(f"\nError initializing Stockfish: {e}\nAnalysis disabled."); engine = None
    input("Press Enter to start the game...")

    # --- Game Loop ---
    while True:
        # Display board, potentially meter/plot based on meter_visible flag
        display_game(board, meter_visible, eval_history, last_raw_score,
                     last_move=last_move_made, analysis_text=one_off_analysis_text, message=message)
        message = None             # Clear message
        one_off_analysis_text = None # Clear one-off analysis text

        # Check for game over *before* asking for input
        if board.is_game_over(claim_draw=True):
            message = "Game Over!"
            display_game(board, meter_visible, eval_history, last_raw_score, last_move=last_move_made, message=message) # Show final state
            break

        # Prompt for move/command
        prompt = (f"Move (UCI/SAN) | Cmds: meter, analyze, history, undo, fen, pos, result, quit: ")
        user_input = input(prompt).strip()
        cmd = user_input.lower()

        # --- Handle Commands ---
        if cmd == 'quit':
            print("Exiting game."); break
        elif cmd == 'meter':
            meter_visible = not meter_visible
            message = f"Evaluation Meter & Plot {'ENABLED' if meter_visible else 'DISABLED'}."
            if meter_visible and not eval_history and not board.move_stack: # Analyze initial pos if meter turned on at start
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 if raw_score is not None:
                     eval_history.append((0, score_to_numerical(raw_score)))
                     last_raw_score = raw_score
            elif meter_visible and eval_history and eval_history[-1][0] != len(board.move_stack): # Re-analyze if history doesn't match current ply
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 if raw_score is not None:
                     # Don't add duplicates, just update last_raw_score for the bar
                     last_raw_score = raw_score

            last_move_made = None # Clear highlight after command
            continue # Redisplay
        elif cmd == 'analyze':
            best_move_san, score_str, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
            if best_move_san:
                one_off_analysis_text = f"Engine Suggestion: {best_move_san} ({score_str})"
                # Also update history/last score if meter happens to be on
                if meter_visible and raw_score is not None:
                    current_ply = len(board.move_stack)
                    numerical_score = score_to_numerical(raw_score)
                    # Avoid duplicate entries for the same ply
                    if not eval_history or eval_history[-1][0] != current_ply:
                        eval_history.append((current_ply, numerical_score))
                    else: # Update score if re-analyzing same position
                         eval_history[-1] = (current_ply, numerical_score)
                    last_raw_score = raw_score
            else:
                message = score_str # Display error msg
            last_move_made = None
            continue
        elif cmd == 'history':
            display_history(board); input("Press Enter to return...")
            last_move_made = None
            continue
        elif cmd == 'undo':
            if board.move_stack:
                try:
                    board.pop()
                    message = "Last move undone."
                    last_move_made = board.peek() if board.move_stack else None
                    # --- Remove last entry from eval history ---
                    if eval_history:
                        eval_history.pop()
                    # Update last_raw_score to the new latest score in history
                    if eval_history:
                        # Need to re-create a Score object approx, or just re-analyze? Re-analyze is easier.
                        if meter_visible:
                           _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                           last_raw_score = raw_score
                        else:
                           last_raw_score = None # Clear score if meter is off
                    else:
                        last_raw_score = None # History is empty
                except IndexError: message = "Cannot undo."; last_move_made = None
            else: message = "No moves to undo."; last_move_made = None
            one_off_analysis_text = None # Clear any previous one-off analysis
            continue
        elif cmd == 'fen': message = f"FEN: {board.fen()}"; last_move_made = None; continue
        elif cmd == 'pos': message = f"Start Pos: {start_pos_num} | FEN: {start_fen}"; last_move_made = None; continue
        elif cmd == 'result':
             if board.is_game_over(claim_draw=True):
                 res = board.outcome(); term = res.termination.name.replace('_',' ').title()
                 message = f"Result: {res.result()} ({term})"
             else: message = "Game in progress."
             last_move_made = None; continue

        # --- Process Move Input ---
        move = None
        try: move = board.parse_uci(user_input)
        except ValueError:
            try: move = board.parse_san(user_input)
            except ValueError as e:
                 err = str(e).lower(); msg_key = "Ambiguous" if "ambiguous" in err else "Illegal" if "illegal" in err else "Invalid"
                 message = f"{msg_key} move: '{user_input}'."; last_move_made = None; continue

        if move in board.legal_moves:
            board.push(move)
            last_move_made = move
            message = None
            one_off_analysis_text = None # Clear previous one-off

            # --- Auto-analyze if meter is visible ---
            if meter_visible:
                _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                if raw_score is not None:
                    current_ply = len(board.move_stack)
                    eval_history.append((current_ply, score_to_numerical(raw_score)))
                    last_raw_score = raw_score
                else:
                    # If analysis fails, maybe append previous score or 0? Let's append 0 for now.
                    current_ply = len(board.move_stack)
                    eval_history.append((current_ply, 0))
                    # Keep the previous last_raw_score for the bar? Or set to None? Let's keep it.
                    # last_raw_score = None # Or keep the old one?

        else: # Illegal move
            parsed_as = user_input
            try: parsed_as = board.san(move)
            except: pass # Ignore errors here
            reason = "puts king in check" if board.is_pseudo_legal(move) else "not legal"
            message = f"Illegal move: {user_input} (parsed {parsed_as}) - {reason}."
            last_move_made = None

    # --- Game Over Sequence ---
    print("\n" + "="*30 + "\n           GAME OVER\n" + "="*30)
    outcome = board.outcome(claim_draw=True)
    if outcome: print(f"\nResult: {outcome.result()} ({outcome.termination.name.replace('_',' ').title()})")
    else: print(f"\nResult: {board.result(claim_draw=True)}") # Fallback
    display_history(board)

    # --- Cleanup ---
    if engine:
        try: engine.quit(); print("\nStockfish engine closed.")
        except: print("\nStockfish engine already closed or error.")
    # Close any lingering matplotlib plots (belt-and-suspenders)
    plt.close('all')

# --- Start the Game ---
play_chess960()
```

```python
import chess
import chess.svg
import chess.engine
import random
import os
import io
import base64
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output, SVG, HTML

# Configuration
STOCKFISH_PATH = "./stockfish/stockfish-macos-m1-apple-silicon"
ANALYSIS_TIME_LIMIT = 1.0
EVAL_BAR_HEIGHT = 350
EVAL_BAR_WIDTH = 20
EVAL_CLAMP_LIMIT = 800
PLOT_HEIGHT = 150
PLOT_WIDTH = EVAL_BAR_HEIGHT + EVAL_BAR_WIDTH + 10
MATE_SCORE_PLOT_VALUE = 15000

# Helper Functions
def get_castling_rights_str(board):
    rights = []
    if board.has_kingside_castling_rights(chess.WHITE): rights.append("K")
    if board.has_queenside_castling_rights(chess.WHITE): rights.append("Q")
    if board.has_kingside_castling_rights(chess.BLACK): rights.append("k")
    if board.has_queenside_castling_rights(chess.BLACK): rights.append("q")
    return "".join(rights) if rights else "-"

def score_to_white_percentage(score, clamp_limit=EVAL_CLAMP_LIMIT):
    if score is None: return None
    pov_score = score.white()
    if pov_score.is_mate():
        mate_val = pov_score.mate()
        return 100 if mate_val is not None and mate_val > 0 else 0
    cp = pov_score.score(mate_score=clamp_limit * 2)
    if cp is None: return None
    clamped_cp = max(-clamp_limit, min(clamp_limit, cp))
    normalized = (clamped_cp + clamp_limit) / (2 * clamp_limit)
    return normalized * 100

def score_to_numerical(score):
    if score is None: return 0
    pov_score = score.white()
    if pov_score.is_mate():
        mate_val = pov_score.mate()
        return MATE_SCORE_PLOT_VALUE if mate_val > 0 else -MATE_SCORE_PLOT_VALUE
    cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE * 2)
    return cp if cp is not None else 0

def create_eval_bar_html(white_percentage, height=EVAL_BAR_HEIGHT, width=EVAL_BAR_WIDTH):
    if white_percentage is None: return ""
    white_percentage = max(0, min(100, white_percentage))
    container_style = f"width: {width}px; height: {height}px; background-color: #333; border: 1px solid #666; position: relative; overflow: hidden; box-sizing: border-box;"
    white_style = f"width: 100%; height: {white_percentage}%; background-color: white; position: absolute; bottom: 0; left: 0;"
    return f'<div style="{container_style}"><div style="{white_style}"></div></div>'

def create_eval_plot_html(eval_history, plot_width_px=PLOT_WIDTH, plot_height_px=PLOT_HEIGHT):
    """Generates HTML containing a base64 encoded plot of the evaluation history."""
    if not eval_history:
        return "" # Return empty if no history

    plies = [item[0] for item in eval_history]
    scores = [item[1] / 100.0 for item in eval_history] # Convert cp to pawns for display

    # Determine plot y-axis limits dynamically but capped
    max_abs_score = 0
    if scores:
        max_abs_score = max(abs(s) for s in scores if abs(s) < MATE_SCORE_PLOT_VALUE / 100.0) # Ignore obvious mates for limits
        max_abs_score = max(max_abs_score, 2) # Ensure at least +/- 2 range

    # Cap the y-limit, slightly above the clamp limit for the bar
    plot_y_limit = min(max(max_abs_score * 1.2, 2), EVAL_CLAMP_LIMIT / 100.0 * 1.5) # Up to 1.5x bar limit

    # Handle Mate scores for display clipping
    plot_y_limit_final = MATE_SCORE_PLOT_VALUE / 100.0 # Use the large value for actual clipping if mate occurs

    fig, ax = plt.subplots(figsize=(plot_width_px / 80, plot_height_px / 80), dpi=80) # Adjust figsize based on desired pixel dimensions
    ax.plot(plies, scores, marker='.', linestyle='-', color='blue')
    ax.axhline(0, color='black', linewidth=0.5, linestyle='--') # Zero line

    ax.set_xlabel("Ply (Half-moves)")
    ax.set_ylabel("Eval (Pawns)")
    ax.set_title("Evaluation History (White's POV)")
    ax.grid(True, linestyle=':', linewidth=0.5)

    # Set y-limits, clipping extreme mate scores visually
    ax.set_ylim(-plot_y_limit_final, plot_y_limit_final) # Clip extreme values

    # Add text annotations near y=0 for clarity if space permits
    # Add plot limits text
    ax.text(plies[-1] if plies else 0, plot_y_limit * 0.9, f'+{plot_y_limit:.1f}', verticalalignment='top', horizontalalignment='right', fontsize=8, color='grey')
    ax.text(plies[-1] if plies else 0, -plot_y_limit * 0.9, f'-{plot_y_limit:.1f}', verticalalignment='bottom', horizontalalignment='right', fontsize=8, color='grey')

    # Fill the bottom area with white and the top area with black
    ax.fill_between(plies, -plot_y_limit_final, 0, color='white')
    ax.fill_between(plies, 0, plot_y_limit_final, color='black')

    # Save plot to a BytesIO object
    buf = io.BytesIO()
    plt.savefig(buf, format='png', bbox_inches='tight') # Use bbox_inches='tight'
    plt.close(fig) # Close the plot to free memory
    buf.seek(0)

    # Encode image to base64
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    buf.close()

    # Create HTML img tag
    return f'<img src="data:image/png;base64,{img_base64}" alt="Evaluation Plot"/>'


def display_game(board, meter_visible, eval_history, last_raw_score, last_move=None, analysis_text=None, message=None):
    clear_output(wait=True)
    king_square = board.king(board.turn) if board.is_check() else None
    svg_board_str = chess.svg.board(board=board, lastmove=last_move, check=king_square, coordinates=True, size=EVAL_BAR_HEIGHT)
    eval_bar_html = ""
    eval_plot_html = ""
    if meter_visible:
        if last_raw_score:
            white_percentage = score_to_white_percentage(last_raw_score, EVAL_CLAMP_LIMIT)
            eval_bar_html = create_eval_bar_html(white_percentage, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)
        else:
            eval_bar_html = create_eval_bar_html(50, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)
        eval_plot_html = create_eval_plot_html(eval_history, PLOT_WIDTH, PLOT_HEIGHT)
    combined_html = f"""
    <table style="border-collapse: collapse; border: none;">
        <tr>
            <td style="padding: 0 10px 0 0; vertical-align: top;">{eval_bar_html}</td>
            <td style="padding: 0; vertical-align: top;">{svg_board_str}</td>
        </tr>
    """
    if eval_plot_html:
        combined_html += f"""
        <tr>
            <td colspan="2" style="padding-top: 10px; text-align: left;">{eval_plot_html}</td>
        </tr>
        """
    combined_html += "</table>"
    display(HTML(combined_html))
    print("-" * 30)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.is_check(): print("!! CHECK !!")
    if last_move:
        try: print(f"Last move: {board.san(last_move)}")
        except Exception as e: print(f"Last move: {last_move.uci()} (SAN error: {e})")
    print(f"Move count: {board.fullmove_number}")
    print(f"Castling Rights: {get_castling_rights_str(board)}")
    print(f"Ply: {len(board.move_stack)}")
    if analysis_text:
        print("-" * 30)
        print(analysis_text)
    if message:
        print("-" * 30)
        print(f"*** {message} ***")
    print("-" * 30)

def get_engine_analysis(board, engine, time_limit):
    if not engine:
        return None, "Analysis unavailable (Engine not configured).", None
    try:
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0]
        score = info.get("score")
        if best_move and score is not None:
            pov_score = score.white()
            score_str = ""
            if pov_score.is_mate():
                mate_in = pov_score.mate()
                score_str = f"Mate in {abs(mate_in)}" if mate_in is not None else "Mate (N/A)"
            else:
                cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE * 2)
                score_str = f"Score: {cp / 100.0:.2f}" if cp is not None else "Score: N/A"
            try: best_move_san = board.san(best_move)
            except Exception: best_move_san = best_move.uci()
            return best_move_san, score_str, score
        else:
            return None, "Engine could not provide suggestion.", None
    except Exception as e:
        print(f"Analysis error: {e}")
        return None, f"Analysis error: {e}", None

def display_history(board):
    history_html = "<h3>Move History (SAN)</h3><ol>"
    temp_board = chess.Board(chess960=board.chess960)
    try: temp_board.set_fen(board.root().fen())
    except Exception as e:
        print(f"Error getting start FEN for history: {e}")
        display(HTML("<li>Error generating history</li></ol>"))
        return
    full_move_parts = []
    for i, move in enumerate(board.move_stack):
        try: move_san = temp_board.san(move)
        except Exception: move_san = move.uci()
        if temp_board.turn == chess.WHITE:
             if full_move_parts: history_html += f"<li>{' '.join(full_move_parts)}</li>"
             full_move_parts = [f"{temp_board.fullmove_number}. {move_san}"]
        else:
            full_move_parts.append(move_san)
        try: temp_board.push(move)
        except Exception as e:
            history_html += f"<li>Error processing move {move.uci()}: {e}</li>"
            break
    if full_move_parts: history_html += f"<li>{' '.join(full_move_parts)}</li>"
    if not board.move_stack: history_html += "<li>No moves made yet.</li>"
    history_html += "</ol>"
    display(HTML(history_html))

def play_chess960():
    board = chess.Board(chess960=True)
    engine = None
    last_move_made = None
    message = None
    start_pos_num = -1
    meter_visible = False
    eval_history = []
    last_raw_score = None
    one_off_analysis_text = None
    while True:
        clear_output()
        pos_choice = input("Enter Chess960 position number (0-959), 'random', or leave blank for random: ").strip().lower()
        if not pos_choice or pos_choice == 'random':
            start_pos_num = random.randint(0, 959)
            break
        else:
            try:
                start_pos_num = int(pos_choice)
                if 0 <= start_pos_num <= 959: break
                else: print("Position number must be between 0 and 959."); input("Press Enter...")
            except ValueError: print("Invalid input."); input("Press Enter...")
    board.set_chess960_pos(start_pos_num)
    start_fen = board.fen()
    print(f"Starting Chess960 Position {start_pos_num}")
    print(f"Starting FEN: {start_fen}")
    input("Press Enter to continue...")
    try:
        if STOCKFISH_PATH and os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            try: engine.configure({"UCI_Chess960": True}); print("Set UCI_Chess960 option.")
            except chess.engine.EngineError: print("UCI_Chess960 option not supported (usually ok).")
            print(f"Stockfish engine loaded: {STOCKFISH_PATH}")
        else: print(f"Warning: Stockfish not found ('{STOCKFISH_PATH}'). Analysis disabled.")
    except Exception as e: print(f"\nError initializing Stockfish: {e}\nAnalysis disabled."); engine = None
    input("Press Enter to start the game...")
    while True:
        display_game(board, meter_visible, eval_history, last_raw_score, last_move=last_move_made, analysis_text=one_off_analysis_text, message=message)
        message = None
        one_off_analysis_text = None
        if board.is_game_over(claim_draw=True):
            message = "Game Over!"
            display_game(board, meter_visible, eval_history, last_raw_score, last_move=last_move_made, message=message)
            break
        prompt = (f"Move (UCI/SAN) | Cmds: meter, analyze, history, undo, fen, pos, result, quit: ")
        user_input = input(prompt).strip()
        cmd = user_input.lower()
        if cmd == 'quit':
            print("Exiting game."); break
        elif cmd == 'meter':
            meter_visible = not meter_visible
            message = f"Evaluation Meter & Plot {'ENABLED' if meter_visible else 'DISABLED'}."
            if meter_visible and not eval_history and not board.move_stack:
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 if raw_score is not None:
                     eval_history.append((0, score_to_numerical(raw_score)))
                     last_raw_score = raw_score
            elif meter_visible and eval_history and eval_history[-1][0] != len(board.move_stack):
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 if raw_score is not None:
                     last_raw_score = raw_score
            last_move_made = None
            continue
        elif cmd == 'analyze':
            best_move_san, score_str, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
            if best_move_san:
                one_off_analysis_text = f"Engine Suggestion: {best_move_san} ({score_str})"
                if meter_visible and raw_score is not None:
                    current_ply = len(board.move_stack)
                    numerical_score = score_to_numerical(raw_score)
                    if not eval_history or eval_history[-1][0] != current_ply:
                        eval_history.append((current_ply, numerical_score))
                    else:
                         eval_history[-1] = (current_ply, numerical_score)
                    last_raw_score = raw_score
            else:
                message = score_str
            last_move_made = None
            continue
        elif cmd == 'history':
            display_history(board); input("Press Enter to return...")
            last_move_made = None
            continue
        elif cmd == 'undo':
            if board.move_stack:
                try:
                    board.pop()
                    message = "Last move undone."
                    last_move_made = board.peek() if board.move_stack else None
                    if eval_history:
                        eval_history.pop()
                    if eval_history:
                        if meter_visible:
                           _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                           last_raw_score = raw_score
                        else:
                           last_raw_score = None
                    else:
                        last_raw_score = None
                except IndexError: message = "Cannot undo."; last_move_made = None
            else: message = "No moves to undo."; last_move_made = None
            one_off_analysis_text = None
            continue
        elif cmd == 'fen': message = f"FEN: {board.fen()}"; last_move_made = None; continue
        elif cmd == 'pos': message = f"Start Pos: {start_pos_num} | FEN: {start_fen}"; last_move_made = None; continue
        elif cmd == 'result':
             if board.is_game_over(claim_draw=True):
                 res = board.outcome(); term = res.termination.name.replace('_',' ').title()
                 message = f"Result: {res.result()} ({term})"
             else: message = "Game in progress."
             last_move_made = None; continue
        move = None
        try: move = board.parse_uci(user_input)
        except ValueError:
            try: move = board.parse_san(user_input)
            except ValueError as e:
                 err = str(e).lower(); msg_key = "Ambiguous" if "ambiguous" in err else "Illegal" if "illegal" in err else "Invalid"
                 message = f"{msg_key} move: '{user_input}'."; last_move_made = None; continue
        if move in board.legal_moves:
            board.push(move)
            last_move_made = move
            message = None
            one_off_analysis_text = None
            if meter_visible:
                _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                if raw_score is not None:
                    current_ply = len(board.move_stack)
                    eval_history.append((current_ply, score_to_numerical(raw_score)))
                    last_raw_score = raw_score
                else:
                    current_ply = len(board.move_stack)
                    eval_history.append((current_ply, 0))
        else:
            parsed_as = user_input
            try: parsed_as = board.san(move)
            except: pass
            reason = "puts king in check" if board.is_pseudo_legal(move) else "not legal"
            message = f"Illegal move: {user_input} (parsed {parsed_as}) - {reason}."
            last_move_made = None
    print("\n" + "="*30 + "\n           GAME OVER\n" + "="*30)
    outcome = board.outcome(claim_draw=True)
    if outcome: print(f"\nResult: {outcome.result()} ({outcome.termination.name.replace('_',' ').title()})")
    else: print(f"\nResult: {board.result(claim_draw=True)}")
    display_history(board)
    if engine:
        try: engine.quit(); print("\nStockfish engine closed.")
        except: print("\nStockfish engine already closed or error.")
    plt.close('all')

play_chess960()
```

```python
import chess
import chess.svg
import chess.engine
import random
import os
import io
import base64
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, clear_output, SVG, HTML
import matplotlib
matplotlib.use('Agg') # Use Agg backend for non-interactive plotting

# Configuration
STOCKFISH_PATH = "./stockfish/stockfish-macos-m1-apple-silicon" # CHANGE THIS TO YOUR STOCKFISH PATH if needed
ANALYSIS_TIME_LIMIT = 0.5 # Reduced for faster updates, increase if needed
EVAL_BAR_HEIGHT = 350
EVAL_BAR_WIDTH = 20
EVAL_CLAMP_LIMIT = 800 # Centipawn limit for eval bar scaling (approx +/- 8 pawns)
PLOT_HEIGHT = 150
PLOT_WIDTH = EVAL_BAR_HEIGHT + EVAL_BAR_WIDTH + 10 # Make plot width match board + bar width
MATE_SCORE_PLOT_VALUE = 15000 # A large centipawn value assigned for mates in percentage calc

# Helper Functions
def get_castling_rights_str(board):
    rights = []
    if board.has_kingside_castling_rights(chess.WHITE): rights.append("K")
    if board.has_queenside_castling_rights(chess.WHITE): rights.append("Q")
    if board.has_kingside_castling_rights(chess.BLACK): rights.append("k")
    if board.has_queenside_castling_rights(chess.BLACK): rights.append("q")
    return "".join(rights) if rights else "-"

def score_to_white_percentage(score, clamp_limit=EVAL_CLAMP_LIMIT):
    """Converts a chess.engine.Score object to White's win percentage (0-100)."""
    if score is None: return None
    pov_score = score.white() # Get score from White's perspective

    if pov_score.is_mate():
        mate_val = pov_score.mate()
        # Mate for white is 100%, mate for black is 0%
        return 100.0 if mate_val is not None and mate_val > 0 else 0.0

    # Use mate_score when getting cp score to handle potential non-mate scores near mate
    cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE)
    if cp is None: return None # Should not happen if not mate, but check anyway

    # Clamp the centipawn score to the defined limits
    clamped_cp = max(-clamp_limit, min(clamp_limit, cp))

    # Normalize the clamped score to a 0-1 range
    # -clamp_limit maps to 0, +clamp_limit maps to 1
    normalized = (clamped_cp + clamp_limit) / (2 * clamp_limit)

    # Convert to percentage
    return normalized * 100.0

# Removed score_to_numerical as history now stores percentages directly for the plot

def create_eval_bar_html(white_percentage, height=EVAL_BAR_HEIGHT, width=EVAL_BAR_WIDTH):
    """Creates the HTML for the vertical evaluation bar."""
    if white_percentage is None:
        # Default to 50% if no score is available
        white_percentage = 50.0
    white_percentage = max(0, min(100, white_percentage)) # Ensure percentage is within 0-100
    container_style = f"width: {width}px; height: {height}px; background-color: #333; border: 1px solid #666; position: relative; overflow: hidden; box-sizing: border-box;"
    # White part fills from the bottom
    white_style = f"width: 100%; height: {white_percentage}%; background-color: white; position: absolute; bottom: 0; left: 0;"
    # Black part fills the rest from the top (implicitly by the container background)
    return f'<div style="{container_style}"><div style="{white_style}"></div></div>'

def create_eval_plot_html(eval_history_percent, plot_width_px=PLOT_WIDTH, plot_height_px=PLOT_HEIGHT):
    """Generates HTML containing a base64 encoded area plot of the win probability history."""
    if not eval_history_percent or len(eval_history_percent) < 2:
        # Need at least two points to draw an area
        return "" # Return empty if not enough history

    plies = [item[0] for item in eval_history_percent]
    percentages = [item[1] if item[1] is not None else 50.0 for item in eval_history_percent] # Use 50% for any None values

    # --- Plotting ---
    fig, ax = plt.subplots(figsize=(plot_width_px / 80, plot_height_px / 80), dpi=80) # Adjust figsize based on desired pixel dimensions

    # Set background color to black
    fig.patch.set_facecolor('black')
    ax.set_facecolor('black')

    # Create the area chart
    # Fill between the percentage line and y=0 (bottom of the plot)
    ax.fill_between(plies, percentages, color='white', alpha=0.9) # Fill with white

    # Add the orange horizontal line at 50%
    ax.axhline(50, color='orange', linestyle='-', linewidth=1.5, label='50% Line')

    # --- Styling ---
    ax.set_xlabel("Ply (Half-moves)", color='grey')
    ax.set_ylabel("White Win Probability (%)", color='grey')
    ax.set_title("Win Probability History", color='white') # White title for contrast

    # Set plot limits
    ax.set_xlim(0, max(plies)) # X-axis starts at ply 0
    ax.set_ylim(0, 100)        # Y-axis is percentage 0-100

    # Customize ticks and grid for visibility on black background
    ax.tick_params(axis='x', colors='grey')
    ax.tick_params(axis='y', colors='grey')
    ax.grid(True, linestyle=':', linewidth=0.5, color='dimgrey') # Dark grey grid

    # Customize spines (plot border) color
    ax.spines['top'].set_color('grey')
    ax.spines['bottom'].set_color('grey')
    ax.spines['left'].set_color('grey')
    ax.spines['right'].set_color('grey')

    # --- Save plot to memory ---
    buf = io.BytesIO()
    # Use bbox_inches='tight' to prevent labels being cut off
    # Increase padding slightly if needed: pad_inches=0.1
    plt.savefig(buf, format='png', bbox_inches='tight', facecolor=fig.get_facecolor(), edgecolor='none', pad_inches=0.05)
    plt.close(fig) # Close the plot to free memory
    buf.seek(0)

    # Encode image to base64
    img_base64 = base64.b64encode(buf.read()).decode('utf-8')
    buf.close()

    # Create HTML img tag
    return f'<img src="data:image/png;base64,{img_base64}" alt="Evaluation Plot"/>'


def display_game(board, meter_visible, eval_history_percent, last_raw_score, last_move=None, analysis_text=None, message=None):
    """Displays the board, eval bar, eval plot, and game info."""
    clear_output(wait=True) # Clears the previous output in Jupyter/Colab

    # Get board SVG
    king_square = board.king(board.turn) if board.is_check() else None
    svg_board_str = chess.svg.board(board=board, lastmove=last_move, check=king_square, coordinates=True, size=EVAL_BAR_HEIGHT)

    # Initialize HTML elements
    eval_bar_html = ""
    eval_plot_html = ""
    white_percentage_for_bar = None

    # Calculate percentage for the eval bar (uses the most recent raw score)
    if last_raw_score:
        white_percentage_for_bar = score_to_white_percentage(last_raw_score, EVAL_CLAMP_LIMIT)

    # Generate HTML only if meter is visible
    if meter_visible:
        # Create eval bar (always show, defaults to 50% if no score)
        eval_bar_html = create_eval_bar_html(white_percentage_for_bar, EVAL_BAR_HEIGHT, EVAL_BAR_WIDTH)
        # Create eval plot (requires history)
        eval_plot_html = create_eval_plot_html(eval_history_percent, PLOT_WIDTH, PLOT_HEIGHT)

    # Combine elements into an HTML table for layout
    combined_html = f"""
    <table style="border-collapse: collapse; border: none;">
        <tr>
            <td style="padding: 0 10px 0 0; vertical-align: top;">{eval_bar_html}</td>
            <td style="padding: 0; vertical-align: top;">{svg_board_str}</td>
        </tr>
    """
    # Add the plot row only if it was generated
    if eval_plot_html:
        combined_html += f"""
        <tr>
            <td colspan="2" style="padding-top: 10px; text-align: left;">{eval_plot_html}</td>
        </tr>
        """
    combined_html += "</table>"

    # Display the HTML
    display(HTML(combined_html))

    # Print game status information below the board/plot
    print("-" * 30)
    turn = "White" if board.turn == chess.WHITE else "Black"
    print(f"Turn: {turn}")
    if board.is_check(): print("!! CHECK !!")
    if last_move:
        try: print(f"Last move: {board.san(last_move)}")
        except Exception as e: print(f"Last move: {last_move.uci()} (SAN error: {e})") # Fallback to UCI
    print(f"Move count: {board.fullmove_number}")
    print(f"Castling Rights: {get_castling_rights_str(board)}")
    print(f"Ply: {len(board.move_stack)}") # Current ply (number of half-moves)

    # Display analysis text if available (e.g., from 'analyze' command)
    if analysis_text:
        print("-" * 30)
        print(analysis_text)

    # Display general messages (e.g., errors, confirmations)
    if message:
        print("-" * 30)
        print(f"*** {message} ***")
    print("-" * 30)

def get_engine_analysis(board, engine, time_limit):
    """Runs engine analysis and returns best move, score string, and raw score object."""
    if not engine:
        return None, "Analysis unavailable (Engine not configured).", None
    try:
        # Run analysis
        info = engine.analyse(board, chess.engine.Limit(time=time_limit))
        best_move = info.get("pv", [None])[0] # Principal variation's first move
        score = info.get("score")             # Score object

        if best_move and score is not None:
            pov_score = score.pov(board.turn) # Score from the current player's perspective
            score_str = ""
            if pov_score.is_mate():
                mate_in = pov_score.mate()
                # Make sure mate is reported correctly (+ for winning, - for losing)
                score_str = f"Mate in {abs(mate_in)}" if mate_in is not None else "Mate (N/A)"
            else:
                # Convert cp score, handle potential None
                cp = pov_score.score(mate_score=MATE_SCORE_PLOT_VALUE * 2) # Provide large mate score if needed
                score_str = f"Score: {cp / 100.0:+.2f}" if cp is not None else "Score: N/A" # Show sign

            # Try to get SAN notation, fall back to UCI
            try: best_move_san = board.san(best_move)
            except Exception: best_move_san = best_move.uci()

            # Return SAN move, score string (for display), and the raw score object (for calculations)
            return best_move_san, score_str, score # Return original score (needed for white % calc)
        else:
            return None, "Engine could not provide suggestion.", None
    except Exception as e:
        print(f"Analysis error: {e}")
        # Return original exception message
        return None, f"Analysis error: {e}", None

def display_history(board):
    """Displays the game's move history in SAN."""
    history_html = "<h3>Move History (SAN)</h3><ol>"
    # Create a temporary board to replay moves for SAN generation
    temp_board = chess.Board(chess960=board.chess960)
    try:
        # Set start position from the root node
        temp_board.set_fen(board.root().fen())
    except Exception as e:
        print(f"Error getting start FEN for history: {e}")
        display(HTML("<li>Error generating history</li></ol>"))
        return

    full_move_parts = [] # To store parts of a full move (e.g., "1. e4 e5")
    # Iterate through the move stack of the main board
    for i, move in enumerate(board.move_stack):
        try:
            # Get Standard Algebraic Notation for the move
            move_san = temp_board.san(move)
        except Exception:
            # Fallback to UCI if SAN fails (should be rare)
            move_san = move.uci()

        # Group moves by full move number
        if temp_board.turn == chess.WHITE:
             # If we have parts from the previous move (Black's move), add the list item
             if full_move_parts:
                 history_html += f"<li>{' '.join(full_move_parts)}</li>"
             # Start a new list item with the move number
             full_move_parts = [f"{temp_board.fullmove_number}. {move_san}"]
        else:
            # Add Black's move to the current list item parts
            full_move_parts.append(move_san)

        # Make the move on the temporary board to keep track
        try:
            temp_board.push(move)
        except Exception as e:
            # If pushing fails, log error and stop
            history_html += f"<li>Error processing move {move.uci()}: {e}</li>"
            break # Stop processing further moves

    # Add the last collected move parts if any exist
    if full_move_parts:
        history_html += f"<li>{' '.join(full_move_parts)}</li>"

    # Handle case where no moves have been made
    if not board.move_stack:
        history_html += "<li>No moves made yet.</li>"

    history_html += "</ol>"
    display(HTML(history_html)) # Display the generated HTML list

def play_chess960():
    """Main function to run the Chess960 game loop."""
    board = chess.Board(chess960=True) # Initialize board for Chess960
    engine = None
    last_move_made = None
    message = None
    start_pos_num = -1
    meter_visible = False # Eval meter/plot visibility flag
    eval_history_percent = [] # Stores tuples of (ply, white_percentage)
    last_raw_score = None # Store the last raw score object for the eval bar
    one_off_analysis_text = None # For displaying temporary analysis results

    # --- Get Starting Position ---
    while True:
        clear_output()
        pos_choice = input("Enter Chess960 position number (0-959), 'random', or leave blank for random: ").strip().lower()
        if not pos_choice or pos_choice == 'random':
            start_pos_num = random.randint(0, 959)
            break
        else:
            try:
                start_pos_num = int(pos_choice)
                if 0 <= start_pos_num <= 959:
                    break
                else:
                    print("Position number must be between 0 and 959.")
                    input("Press Enter...") # Pause for user
            except ValueError:
                print("Invalid input.")
                input("Press Enter...") # Pause for user

    # Set the chosen Chess960 position
    board.set_chess960_pos(start_pos_num)
    start_fen = board.fen() # Store the starting FEN
    print(f"Starting Chess960 Position {start_pos_num}")
    print(f"Starting FEN: {start_fen}")
    input("Press Enter to continue...")

    # --- Initialize Engine ---
    try:
        if STOCKFISH_PATH and os.path.exists(STOCKFISH_PATH):
            engine = chess.engine.SimpleEngine.popen_uci(STOCKFISH_PATH)
            # Try setting Chess960 mode (might fail on some Stockfish versions/builds, but often works anyway)
            try:
                engine.configure({"UCI_Chess960": True})
                print("Set UCI_Chess960 option.")
            except (chess.engine.EngineError, KeyError, TypeError): # Catch various config errors
                 print("UCI_Chess960 option not supported or failed (usually ok).")
            print(f"Stockfish engine loaded: {STOCKFISH_PATH}")
        else:
            print(f"Warning: Stockfish not found or path incorrect ('{STOCKFISH_PATH}'). Analysis disabled.")
    except Exception as e:
        print(f"\nError initializing Stockfish: {e}\nAnalysis disabled.")
        engine = None # Ensure engine is None if init fails
    input("Press Enter to start the game...")

    # --- Game Loop ---
    while True:
        # Display current game state
        display_game(board, meter_visible, eval_history_percent, last_raw_score, last_move=last_move_made, analysis_text=one_off_analysis_text, message=message)
        message = None # Clear message after display
        one_off_analysis_text = None # Clear analysis text

        # Check for game over conditions
        if board.is_game_over(claim_draw=True):
            message = "Game Over!"
            display_game(board, meter_visible, eval_history_percent, last_raw_score, last_move=last_move_made, message=message)
            break # Exit game loop

        # --- User Input ---
        prompt = (f"Move (UCI/SAN) | Cmds: meter, analyze, history, undo, fen, pos, result, quit: ")
        user_input = input(prompt).strip()
        cmd = user_input.lower() # Use lowercase for command matching

        # --- Command Handling ---
        if cmd == 'quit':
            print("Exiting game.")
            break
        elif cmd == 'meter':
            meter_visible = not meter_visible
            message = f"Evaluation Meter & Plot {'ENABLED' if meter_visible else 'DISABLED'}."
            # If enabling meter and no history exists yet, analyze current position
            if meter_visible and not eval_history_percent:
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 last_raw_score = raw_score # Store for bar
                 percent = score_to_white_percentage(raw_score, EVAL_CLAMP_LIMIT)
                 current_ply = len(board.move_stack)
                 # Add initial point (or 50% if analysis failed)
                 eval_history_percent.append((current_ply, percent if percent is not None else 50.0))
            elif meter_visible and eval_history_percent:
                 # If meter was off, ensure the last_raw_score is updated
                 _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                 last_raw_score = raw_score # Update score for eval bar display
            last_move_made = None # Don't highlight move on command toggle
            continue # Redraw with updated meter state

        elif cmd == 'analyze':
            best_move_san, score_str, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT * 2) # Allow more time for manual analysis
            if best_move_san:
                one_off_analysis_text = f"Engine Suggestion: {best_move_san} ({score_str})"
                last_raw_score = raw_score # Update score for bar
                # If meter is on, update the plot history for the current ply
                if meter_visible and raw_score is not None:
                    current_ply = len(board.move_stack)
                    percent = score_to_white_percentage(raw_score, EVAL_CLAMP_LIMIT)
                    if percent is not None:
                        # If history exists and last entry is for the same ply, update it
                        if eval_history_percent and eval_history_percent[-1][0] == current_ply:
                            eval_history_percent[-1] = (current_ply, percent)
                        else: # Otherwise, append new entry
                            eval_history_percent.append((current_ply, percent))
            else:
                message = score_str # Display error/reason if analysis failed
            last_move_made = None # Don't highlight move
            continue # Redraw with analysis result

        elif cmd == 'history':
            display_history(board)
            input("Press Enter to return...") # Pause to view history
            last_move_made = None # Don't highlight move
            continue # Redraw game

        elif cmd == 'undo':
            if board.move_stack:
                try:
                    board.pop() # Undo the last move
                    message = "Last move undone."
                    last_move_made = board.peek() if board.move_stack else None # Update last move highlight
                    # Remove last entry from plot history if it exists
                    if eval_history_percent:
                        eval_history_percent.pop()
                    # Re-analyze the new current position to update the score bar/plot point
                    if meter_visible:
                        _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                        last_raw_score = raw_score # Update score for bar
                        # Note: We don't add the undone state back to plot history here,
                        # it reflects the state *after* the undo.
                    else:
                       last_raw_score = None
                except IndexError:
                    message = "Cannot undo." # Should not happen if move_stack check passed
                    last_move_made = None
            else:
                message = "No moves to undo."
                last_move_made = None
            one_off_analysis_text = None # Clear any previous analysis text
            continue # Redraw

        elif cmd == 'fen':
            message = f"FEN: {board.fen()}"
            last_move_made = None
            continue # Redraw with FEN message
        elif cmd == 'pos':
            message = f"Start Pos: {start_pos_num} | Start FEN: {start_fen}"
            last_move_made = None
            continue # Redraw with start position info
        elif cmd == 'result':
             if board.is_game_over(claim_draw=True):
                 res = board.outcome(claim_draw=True)
                 term = res.termination.name.replace('_',' ').title() if res else "N/A"
                 message = f"Result: {res.result() if res else board.result(claim_draw=True)} ({term})"
             else:
                 message = "Game in progress."
             last_move_made = None
             continue # Redraw with result message

        # --- Move Handling ---
        move = None
        try:
            # Try parsing as UCI first (e.g., "e2e4")
            move = board.parse_uci(user_input)
        except ValueError:
            try:
                # If UCI fails, try parsing as SAN (e.g., "Nf3", "O-O")
                move = board.parse_san(user_input)
            except ValueError as e:
                 # Determine if ambiguous, illegal, or invalid based on error
                 err_lower = str(e).lower()
                 if "ambiguous" in err_lower: msg_key = "Ambiguous"
                 elif "illegal" in err_lower: msg_key = "Illegal"
                 else: msg_key = "Invalid"
                 message = f"{msg_key} move: '{user_input}'. Error: {e}"
                 last_move_made = None
                 continue # Ask for input again

        # Check if the parsed move is legal
        if move in board.legal_moves:
            board.push(move) # Make the move
            last_move_made = move # Store for highlighting
            message = None       # Clear any previous error messages
            one_off_analysis_text = None # Clear analysis text

            # If meter is visible, get analysis and update history
            if meter_visible:
                _, _, raw_score = get_engine_analysis(board, engine, ANALYSIS_TIME_LIMIT)
                last_raw_score = raw_score # Update score for bar
                percent = score_to_white_percentage(raw_score, EVAL_CLAMP_LIMIT)
                current_ply = len(board.move_stack)
                # Append new percentage (or 50% if analysis failed)
                eval_history_percent.append((current_ply, percent if percent is not None else 50.0))

        else:
            # Move was parsed but is not legal
            parsed_as = user_input
            try: parsed_as = board.san(move) # Show SAN if possible
            except: pass # Keep original input if SAN fails
            # Check if it was pseudo-legal (e.g., moves into check)
            reason = "puts king in check" if board.is_pseudo_legal(move) else "not a legal move"
            message = f"Illegal move: {user_input} (parsed as {parsed_as}) - {reason}."
            last_move_made = None
            continue # Ask for input again

    # --- Post-Game ---
    print("\n" + "="*30 + "\n           GAME OVER\n" + "="*30)
    outcome = board.outcome(claim_draw=True)
    if outcome:
        term_name = outcome.termination.name.replace('_',' ').title()
        print(f"\nResult: {outcome.result()} ({term_name})")
    else:
        # Fallback if outcome somehow not available (shouldn't happen)
        print(f"\nResult: {board.result(claim_draw=True)}")

    # Display final history
    display_history(board)

    # --- Cleanup ---
    if engine:
        try:
            engine.quit()
            print("\nStockfish engine closed.")
        except Exception: # Catch potential errors if engine already closed
            print("\nStockfish engine already closed or error during quit.")
    # Close any remaining matplotlib figures
    plt.close('all')

# --- Run the Game ---
play_chess960()
```
